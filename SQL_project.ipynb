```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('inventory_forecasting.csv')
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
df
```

```python
print(df.duplicated().sum())
df.drop(df[df.duplicated()].index, inplace=True)
```

```python
categorical_cols = ['Store ID', 'Product ID', 'Category', 'Region', 'Weather Condition', 'Holiday/Promotion', 'Seasonality']

for col in categorical_cols:
    # Check if the column's data type is object (which typically holds strings)
    if df[col].dtype == 'object':
        df[col] = df[col].str.strip()
        df[col] = df[col].str.title()
    else:
        # Optional: print a warning or message if a column is skipped
        print(f"Skipping .str operations on column '{col}' as it is not of object dtype.")
```

```python
df['Inventory Turnover'] = df['Units Sold'] / df['Inventory Level']
df
```

```python
df.to_csv('cleaned_inventory_forecasting.csv', index=False)
```
